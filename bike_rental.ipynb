# Predicting Bike Rentals

This notebook contains an analysis of bike rental data obtained from the [UCI Machine Learning Repository](http://archive.ics.uci.edu/ml/datasets/Bike+Sharing+Dataset). The data was compiled by Hadi Fanaee-T and contains hourly counts of bike rentals through the Capital Bikeshare program in Washington DC. Our goal will to build a model which can accurately predict the number of bike rentals at a on a given day and time. The dataset contains information reguarding the weather, day of the week, time, and whether it was a holiday. We will be using root mean squared error to evaluate model performance and our final model will be a Random Forest and achieves a root mean squared error of approximately 80.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Feature Engineering

We'll create some new features to see if that can help with our prediction power. First we'll divide the `hour` column into buckets since it doesn't make sense as a numerical variable and there are too many hours to treat each as its own category.

```python
def assign_label(hour):
    if 6 <= hour < 12:
        return 1
    elif 12 <= hour < 18:
        return 2
    elif 18 <= hour < 24:
        return 3
    elif 0 <= hour < 6:
        return 4
bike_rentals['time_label'] = bike_rentals['hr'].apply(assign_label)
```

Let's now take a look at the `cnt` column after appying a box-cox transformation.

```python
from sklearn.preprocessing import power_transform
bike_rentals_boxcox = bike_rentals.copy()
bike_rentals_boxcox['cnt_boxcox'] = power_transform(bike_rentals[['cnt']], method='box-cox')
```

```python
numeric_cols = ['temp', 'atemp', 'windspeed', 'hum', 'cnt_boxcox']
sns.pairplot(bike_rentals_boxcox[numeric_cols], plot_kws={"s": 10})
```

The data looks much more normal so we'll try this when fitting a linear linear regression.

## Model Selection

In this section we'll look at several different machine learning methods in order to figure out which is the best at prediciton. The three methods we'll try are:

1. Linear Regression
2. Poisson Regression
3. Random Forest

Before we begin building the models we'll prepare the data for modeling. First we'll extract the categorical features and convert them into dummy variables.

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def prepare_data(df, numeric_features, categorical_features):
    '''Extracts requested features, and converts categorical features to dummies to prepare
        for model fitting.'''

    # Convert Categories into dummy variables
    dummies = pd.get_dummies(pd.DataFrame({category: df[category].astype('category') 
                for category in categorical_features}), drop_first=True)

    # Reattach categorical variables (leaving out original categorical variables)
    df = pd.concat([df[numeric_features], dummies], axis=1)
    return df
```

Now we'll split the data into training and test sets. We'll reserve 30% of the data for testing.

```python
# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(bike_rentals.drop(columns=['cnt']), bike_rentals['cnt'], 
                                                    test_size=0.3, random_state=19030)
```

Finally, we'll define a function which takes an estimator and returns the root mean squared error based on the test set. This will make it more convenient for us to try different iterations of our models without writing too much code.

```python
# Define function to automatically evaluate model
def evaluate_model(model, X_train, X_test, y_train, y_test, numeric_features, categorical_features):
    '''Take in model, prepares test and training data, fits model, computes MSE on holdout set.'''
    
    # Clean train and test data
    X_train_clean = prepare_data(X_train.copy(), numeric_features, categorical_features)
    X_test_clean = prepare_data(X_test.copy(), numeric_features, categorical_features)
    
    # Fit model, compute predictions on holdout, and compute MSE
    model.fit(X_train_clean, y_train)
    test_predictions = model.predict(X_test_clean)
    mse = mean_squared_error(y_test, test_predictions)
    return mse
```

### Linear Regression

We'll first try linear regression. As a first pass| we'll only consider the orignal data and use Recursive Feature Elimination to select the best subset of features.

```python
from sklearn.feature_selection import RFECV
from sklearn.linear_model import LinearRegression

# Define numeric and categorical variables
numeric_features = ['temp', 'atemp', 'windspeed', 'hum']
categorical_features = ['hr', 'season', 'mnth', 'workingday', 'weekday', 'holiday', 'weathersit']

# Define model and RFE selector
lm = LinearRegression()
selector_lm = RFECV(estimator=lm, cv=10, scoring='neg_mean_squared_error')

# Fit and evaluate
lm_mse = evaluate_model(selector_lm, X_train, X_test, y_train, y_test, numeric_features, categorical_features)
```

```python
print('The mean squared error for Linear Regression on the test set is: %d' %lm_mse)
```

We've now extablished a baseline for performance. Let's try adding and changing some features. First we'll see what happens if we use the box-cox transformation on the counts. Note that due to some quirks in sklearn we'll have to define our own class to transform the target variable.

```python
from sklearn.compose import TransformedTargetRegressor
from sklearn.preprocessing import PowerTransformer

# Define new transformer class
class MyTransformedTargetRegressor(TransformedTargetRegressor):
    @property
    def feature_importances_(self):
        return self.regressor_.feature_importances_

    @property
    def coef_(self):
        return self.regressor_.coef_


ttr = MyTransformedTargetRegressor(regressor=LinearRegression(), transformer=PowerTransformer(method='yeo-johnson'))
selector_lm = RFECV(estimator=ttr, cv=10, scoring='neg_mean_squared_error')
lm_mse = evaluate_model(selector_lm, X_train, X_test, y_train, y_test, numeric_features, categorical_features)
```

```python
print('The mean squared error for Linear Regression on the test set is: %d' %lm_mse)
```

As we can see the transformation did improve the model performance a bit. Let now see if our `time_label` increases our prediction power.

```python
# Define numeric and categorical variables
numeric_features = ['temp', 'atemp', 'windspeed', 'hum']
categorical_features = ['time_label', 'season', 'mnth', 'workingday', 'weekday', 'holiday', 'weathersit']

# Define RFE selector
selector_lm = RFECV(estimator=ttr, cv=10, scoring='neg_mean_squared_error')

# Fit and evaluate
lm_mse = evaluate_model(selector_lm, X_train, X_test, y_train, y_test, numeric_features, categorical_features)
```

```python
print('The mean squared error for Linear Regression on the test set is: %d' %lm_mse)
```

As we see, we actually lose quite a bit of power from moving the hour variable, so we'll keep it and use the second interation of our model to compare going forward.

### Poisson Regression

Since we're actually working with count data it make improve performance to use Poisson Regression. We'll follow the same process as before except this time we'll stick with `hr` and won't consider a box-cox transformation on the target variable because Poisson Regressions require count data.

```python
from sklearn.linear_model import PoissonRegressor

# Define numeric and categorical variables
numeric_features = ['temp', 'atemp', 'windspeed', 'hum']
categorical_features = ['hr', 'season', 'mnth', 'workingday', 'weekday', 'holiday', 'weathersit']

# Define and fit model
pr = PoissonRegressor(alpha=0, max_iter=500)
selector_poisson = RFECV(estimator=pr, cv=10, scoring='neg_mean_squared_error')
pr_mse = evaluate_model(selector_poisson, X_train, X_test, y_train, y_test, numeric_features, categorical_features)
```

```python
print('The mean squared error for Poisson Regression on the test set is: %d' %pr_mse)
```

Poisson Regression does even better than Linear Regression improving the MSE from 11392 to 10934. We'll not try Random Forests.

### Random Forest

We'll won't tune any hyperparameters at first to get a good baseline for performance.

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor(random_state=1988)
mse_forest = evaluate_model(forest, X_train, X_test, y_train, y_test, numeric_features, categorical_features)
```

```python
print('The mean squared error for the Random Forest on the test set is: %d' %mse_forest)
```

This is already a massive improvement over any method we've previously tried. Let's see if we can improve it even further by tuning the hyper parameters. Note that one can play with the parameters below to hone in on the best hyperparameters. We've already don't this to some extent which is why the ranges of parameters are rather small below. We'll be fitting 50 trees here since it takes a while to fit but once we have our optimal hyperparameters well increase the number of iterations for our final estimator.

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [60, 65, 70],
    'max_features': ['auto', 'sqrt'],
    'min_samples_leaf': [1, 2, 3],
    'min_samples_split': [8, 10, 12]
}
forest_gs = RandomForestRegressor(n_estimators=50, random_state=1988)
selector_forest_gs = GridSearchCV(forest_gs, param_grid=param_grid)
mse_forest_gs = evaluate_model(selector_forest_gs, X_train, X_test, y_train, y_test, numeric_features, categorical_features)
```

```python
print('Best Parameters:')
print(selector_forest_gs.best_params_)
print('The mean squared error for the Random Forest on the test set is: %d' %mse_forest_gs)
```

We'll now put these optimal hyper parameters into a RandomForest with more trees and re-fit.

```python
forest_opt = RandomForestRegressor(n_estimators=500, max_depth=60, max_features='auto', 
                                   min_samples_leaf=1, min_samples_split=8, random_state=1988)
mse_forest_opt = evaluate_model(forest_opt, X_train, X_test, y_train, y_test, numeric_features, categorical_features)
```

```python
print('The mean squared error for the Random Forest on the test set is: %d' %mse_forest_opt)
```

This has the best performance yet and we'll select it as our final model. Taking the square root of the mean square error to get approximately 75. This is approximately the error in a typical estimate.
